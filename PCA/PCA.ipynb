# PCA

```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
import numpy as np
import matplotlib.pyplot as plt
```

## Preprocessing

```python
iris = load_iris()
```

```python
X = iris.data
y = iris.target
labels = iris.target_names
y = np.reshape(y, [-1, 1])
```

```python
scaler = StandardScaler()
scaler.fit(X)
x = scaler.transform(X)
```

### Covariance Matrix

```python
matrix = np.cov(x.T)
print(x.shape)
print(matrix.shape)
```

### Eigenvalues and Eigenvectors

```python
val, vec = np.linalg.eig(matrix)
print(val.shape)
print(vec.shape)
```

```python
print(val)
print(vec)
```

### Data Transformation

```python
Z = np.matmul(x, vec)
print(Z.shape)
print(y.shape)
```

### One dimension

```python
z_one = np.reshape(Z[:, 0], [-1, 1])
plt.figure()

for i in range(4):
    rows = np.where(y == i)[0]
    z_temp = z_one[rows, :]
    plt.scatter(z_temp, np.zeros(z_temp.shape))
    
plt.legend(labels)
```

### Two dimensions

```python
z_one = np.reshape(Z[:, 0:2], [-1, 2])
plt.figure()

for i in range(4):
    rows = np.where(y == i)[0]
    z_temp = z_one[rows, :]
    plt.scatter(z_temp[:, 0], z_temp[:, 1])
    
plt.legend(labels)
```

### Three dimensions

```python
z_one = Z[:, 0:3]
fig = plt.figure()
ax = fig.add_subplot(projection="3d")

for i in range(4):
    rows = np.where(y == i)[0]
    z_temp = z_one[rows, :]
    ax.scatter(z_temp[:, 0], z_temp[:, 1], z_temp[:, 2])
    
plt.legend(labels)
```
